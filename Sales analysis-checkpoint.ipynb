```python
import pandas as pd
import os


df = pd.read_csv(r'Pandas-Data-Science-Tasks-master\SalesAnalysis\Sales_Data\Sales_April_2019.csv')



# to create empty dataframe
all_months_data = pd.DataFrame()


# to list all the month data files
files = [file for file in os.listdir(r'Pandas-Data-Science-Tasks-master\SalesAnalysis\Sales_Data')]
for file in files:
    df = pd.read_csv(r'Pandas-Data-Science-Tasks-master\SalesAnalysis\Sales_Data/' + file)
    all_months_data = pd.concat([all_months_data,df])
    

all_months_data.to_csv(r'Pandas-Data-Science-Tasks-master\SalesAnalysis\all_dat.csv', index = False)      

```

```python
all_data = pd.read_csv(r'Pandas-Data-Science-Tasks-master\SalesAnalysis\all_dat.csv')
all_data
```

## Analysis of data by months

```python
# to detect the NAN in df
nan_df = all_data[all_data.isna().any(axis=1)]
nan_df
```

```python
# to remove 'or' from df
all_data = all_data[all_data['Order Date'].str[0:2] != 'Or']
```

```python
# string to integer
all_data['Quantity Ordered'] = pd.to_numeric(all_data['Quantity Ordered'])
all_data['Price Each'] = pd.to_numeric(all_data['Price Each'])
```

```python
# getting month Order Date
all_data['Month'] = all_data['Order Date'].str[0:2]
all_data['Month'] = all_data['Month'].astype('int32')
all_data.head()
```

```python
# calculating Price
all_data['sales'] = all_data['Quantity Ordered'] * all_data['Price Each']
all_data.head()
```

```python
# Grouping by months
results = all_data.groupby('Month').sum()
```

```python
import matplotlib.pyplot as plt

months = range(1,13)

plt.bar(months,results['sales'])
plt.xticks(months)
plt.xlabel('months')
plt.ylabel('Sales in usd')
plt.show()
```

## Analysis of data to know when to diplay advertises 

```python
all_data['date'] = pd.to_datetime(all_data['Order Date'])
```

```python
from tqdm import tqdm
all_data['Dates'] = tqdm(pd.to_datetime(all_data['date']).dt.date)
all_data['Time'] = tqdm(pd.to_datetime(all_data['date']).dt.time)
```

```python
all_data['hour'] = all_data['date'].dt.hour
```

```python
hours = [hour for hour,df in all_data.groupby('hour')]
plt.xticks(hours)
plt.grid()
plt.plot(hours, all_data.groupby(['hour']).count())
```
